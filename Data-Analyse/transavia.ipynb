```python
transavia = pd.read_csv('../scraping/transaviaApi.csv')
```

```python
dfMonth = transavia['ArrivalDateTime'].apply(lambda i: i.split('-')[1])
transavia.loc[transavia.index, 'month'] = dfMonth
```

```python
depTime = transavia['DepartureDateTime'].apply(lambda i: i.split('T')[1])
transavia.loc[transavia.index, 'depTime'] = depTime

arrivalTime = transavia['ArrivalDateTime'].apply(lambda i: i.split('T')[1])
transavia.loc[transavia.index, 'arrivalTime'] = arrivalTime

DepartureDateTime = transavia['DepartureDateTime'].apply(lambda i: i.split('T')[0])
transavia.loc[transavia.index, 'DepartureDateTime'] = DepartureDateTime

ArrivalDateTime = transavia['ArrivalDateTime'].apply(lambda i: i.split('T')[0])
transavia.loc[transavia.index, 'ArrivalDateTime'] = ArrivalDateTime
```

# Transavia	Brussel	Faro	Gemiddelde prijs	Juni


```python
prijs = transavia.query('DepartureAirport== "BRU" and ArrivalAirport== "FAO" and month == "06"')['BaseFare'].sum()
print(prijs)

tax = transavia.query('DepartureAirport== "BRU" and ArrivalAirport== "FAO" and month == "06"')['TaxSurcharge'].sum()
print(tax)

totaal = prijs + tax
print(totaal)

aantal = transavia.query('DepartureAirport== "BRU" and ArrivalAirport== "FAO" and month == "06"')['BaseFare'].count()
print(aantal)

gemiddelde = totaal / aantal
print(gemiddelde)
```
