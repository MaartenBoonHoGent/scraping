# transavia data transform

```python
depTime = transavia['DepartureDateTime'].apply(lambda i: i.split('T')[1])
transavia.loc[transavia.index, 'depTime'] = depTime

arrivalTime = transavia['ArrivalDateTime'].apply(lambda i: i.split('T')[1])
transavia.loc[transavia.index, 'arrivalTime'] = arrivalTime

departDate = transavia['DepartureDateTime'].apply(lambda i: i.split('T')[0])
transavia.loc[transavia.index, 'DepartureDateTime'] = departDate

arrivalDate = transavia['ArrivalDateTime'].apply(lambda i: i.split('T')[0])
transavia.loc[transavia.index, 'ArrivalDateTime'] = arrivalDate
```

```python
transavia.rename(columns= {"FlightNumber":"flightNumber"},inplace=True)
transavia.rename(columns= {"DepartureAirport":"departAirportCode"},inplace=True)
transavia.rename(columns= {"ArrivalAirport":"arrivalAirportCode"},inplace=True)
transavia.rename(columns= {"DepartureDateTime":"departDate"},inplace=True)
transavia.rename(columns= {"ArrivalDateTime":"arrivalDate"},inplace=True)

```
